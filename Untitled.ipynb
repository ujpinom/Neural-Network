```python
import numpy as np;import matplotlib.pyplot as plt
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
plt.scatter(X,y,edgecolors='black')
plt.title('Caracter√≠stica Polinomial')
```

```python
from sklearn.preprocessing import PolynomialFeatures;from sklearn.linear_model import LinearRegression

plf=PolynomialFeatures(degree=2, include_bias=False)
X_poly=plf.fit_transform(X)
lr=LinearRegression()
lr.fit(X_poly,y)
y_pred=lr.predict(plf.fit_transform(np.arange(-100,100,0.5)[:,np.newaxis]))
plt.plot(np.arange(-100,100,0.5),y_pred,label='Regresi√≥n Polinomial')
plt.legend()
plt.scatter(X,y,edgecolors='black',c='red')
plt.title('Caracter√≠stica Polinomial')
plt.axis([-3,3,-1,11])
```

# Learning Curves




```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
def plot_learning_curves(modelo,X,y,grado=1):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
    error_val,error_train=[],[]
    
    for i in range(1,len(X_train)):
        modelo.fit(X_train[:i],y_train[:i])
        y_pred_train=modelo.predict(X_train[:i])
        y_pred_val=modelo.predict(X_test)
        error_train.append(np.sqrt(mean_squared_error(y_train[:i],y_pred_train)))
        error_val.append(np.sqrt(mean_squared_error(y_test,y_pred_val)))
    
    plt.plot(error_train,"r-+",label='RMSE Conjunto de Entrenamiento',c='red')
    plt.plot(error_val,label='RMSE Conjunto de validaci√≥n',c='blue')
    plt.axis([0,80,0,3])
    plt.legend()
    plt.xlabel("Training set size", fontsize=14) # not shown
    plt.ylabel("RMSE", fontsize=14)
    plt.title(f'Grado:{grado}')
        
    



```

```python
lr2=LinearRegression()

plot_learning_curves(lr2,X,y)
```

###### De la gr√°fica anterior se observa como las dos curvas tratan de converger a medida que el tama√±o de datos del conjunto de entrenamiento aumenta, reduciendo as√≠ el overfitting. Sin embargo, ya sabemos de antemano que los datos fueron generados a partir de una funci√≥n cuadr√°tica por lo que un modelo de regresi√≥n lineal nunca va a tener un fit apropiado, es por ello que el error acumulado RMSE es elevado para ambas curvas.

```python
from sklearn.pipeline import Pipeline

modelo=Pipeline([('pol_features',PolynomialFeatures(degree=2,include_bias=False)),('lr',LinearRegression())])

plot_learning_curves(modelo,X,y,2)
```

##### Se observa como para un polinomio de grado el error acumulado es menor, adem√°s se observa la convergencia de las dos curvas conforme aumenta el tama√±o del conjunto de datos de entrenamiento.

```python
modelo=Pipeline([('pol_features',PolynomialFeatures(degree=10,include_bias=False)),('lr',LinearRegression())])

plot_learning_curves(modelo,X,y,10)
```

##### Para un polinomio de grado 10 se observa un gap entre ambas curvas que no parece desaparecer conforme el tama√±o de datos aumenta, esto es un indicio de overfitting del modelo

```python
def plot_varios_modelos(modelo,X,y,grado_modelo=1):
    plt.scatter(X,y,color='blue',s=10)
    modelo.fit(X,y)
    y_pred=modelo.predict(np.arange(-100,100,0.1)[:,np.newaxis])
    label=f'Grado: {grado}'
    plt.plot(np.arange(-100,100,0.1),y_pred,label=label,linewidth=3)
    plt.axis([-3,3,0,10])
    plt.legend()
    
    
    
```

```python
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
modelo1=Pipeline([('pol_features',PolynomialFeatures(degree=1,include_bias=False)),('st',st),('lr',LinearRegression())])
modelo2=Pipeline([('pol_features',PolynomialFeatures(degree=2,include_bias=False)),('st',st),('lr',LinearRegression())])
modelo3=Pipeline([('pol_features',PolynomialFeatures(degree=10,include_bias=False)),('st',st),('lr',LinearRegression())])
modelo4=Pipeline([('pol_features',PolynomialFeatures(degree=150,include_bias=False)),('st',st),('lr',LinearRegression())])

modelos=[modelo1,modelo2,modelo3,modelo4]
grados=[1,2,10,150]

for modelo,grado in zip(modelos,grados):
    plot_varios_modelos(modelo,X,y,grado_modelo=grado)
```

###### En la gr√°fica anterior se muestra varias curvas de distintos grados. La curva de grado 150 se ve como presenta mayor sensibilidad a peque√±os cambios. Es evidente que este √∫ltimo modelo sufre de overfitting.
