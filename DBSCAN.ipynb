# DBSCAN

```python
from sklearn.datasets import make_moons;import matplotlib.pyplot as plt;import numpy as np
np.random.seed(42)
X,y= make_moons(n_samples=1000,noise=0.05)
plt.figure(figsize=(10,5))
plt.scatter(X[:,0],X[:,1],edgecolors='black')
```

```python
from sklearn.cluster import DBSCAN
dbscan=DBSCAN(eps=0.05,min_samples=5)
dbscan.fit(X)
```

```python
anomalias=X[dbscan.labels_==-1]
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.scatter(X[:,0],X[:,1],edgecolors='black',c=dbscan.labels_)

plt.subplot(122)
plt.scatter(X[:,0],X[:,1],edgecolors='black',c=dbscan.labels_)
plt.scatter(anomalias[:,0],anomalias[:,1],marker='x',color='red',label='Anomalia')
plt.legend()
```

```python
len(np.unique(dbscan.labels_))
```

### En la figura se observa como el modelo encuentra mas clusters de los que realmente hay. En realidad encuentra 8 clusters totales, lo que quiere decir que el valor de eps es demasiado pequeo para este conjunto de datos. Ahora vamos a ver como se comporta el modelo si se aumenta eps.

```python
dbscan=DBSCAN(n_jobs=-1,min_samples=5,eps=0.2)
dbscan.fit(X)

anomalias=X[dbscan.labels_==-1]
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.scatter(X[:,0],X[:,1],edgecolors='black',c=dbscan.labels_)

plt.subplot(122)
plt.scatter(X[:,0],X[:,1],edgecolors='black',c=dbscan.labels_)
plt.scatter(anomalias[:,0],anomalias[:,1],marker='x',color='red',label='Anomalia')
plt.legend()
```

### Con un valor para eps de 0.2 el modelo es capaz de identificar correctamente cada uno de los clusters adems de no considerar ninguna anomala.
