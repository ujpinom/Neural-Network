```python
import numpy as np;
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
import numpy as np; import matplotlib.pyplot as plt;
plt.scatter(X,y)
```

### Conjuntos de datos sobre los que se va a mostrar el funcionamiento de Gradient Boosted Regression Trees


### GBRT funciona de forma similar al AdaBoost, solo que esta vez el modelo subsequente lo entrena sobre los errores de predicción del anterior modelo a este (y-y_pred)

```python
from sklearn.tree import DecisionTreeRegressor;from sklearn.model_selection import train_test_split

tree_reg1 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg1.fit(X, y)


y2 = y - tree_reg1.predict(X)
tree_reg2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg2.fit(X, y2)

y3 = y2 - tree_reg2.predict(X)
tree_reg3 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg3.fit(X, y3)

X_new = np.array([[0.8]])

y_pred = sum(tree.predict(X_new) for tree in (tree_reg1, tree_reg2, tree_reg3))
y_pred
```

### En el anterior script se muestra de forma resumida las bases sobre las cuales se cimienta el GBRT

```python
X_test = np.random.rand(200, 1) - 0.5
y_pred=sum(tree.predict(X_test) for tree in (tree_reg1, tree_reg2, tree_reg3))
```

```python
plt.scatter(X_test,y_pred,color='red',label='Predicción')
plt.scatter(X,y,color='orange',alpha=0.5, label='Entrenamiento')
plt.legend(loc='upper center')
```

## Trade-off entre learning-rate y número de estimadores del modelo


```python
from sklearn.ensemble import GradientBoostingRegressor

modelo1=GradientBoostingRegressor(max_depth=2, n_estimators=10, learning_rate=1, random_state=42)
modelo2=GradientBoostingRegressor(max_depth=2, n_estimators=200, learning_rate=0.1, random_state=42)

modelo1.fit(X,y)
modelo2.fit(X,y)
plt.figure(figsize=(12,4))

plt.subplot(121)
plt.scatter(X,y)
x1 = np.linspace(X.min(), X.max(), 500)[:,np.newaxis]
plt.plot(x1,modelo1.predict(x1),color='red',label='predicción')
plt.title(f'Learning_rate: {0.1}.Estimadores: {3}')
plt.legend()

plt.subplot(122)
plt.scatter(X,y)
plt.plot(x1,modelo2.predict(x1),color='red',label='predicción')
plt.title(f'Learning_rate: {0.1}.Estimadores: {200}')
plt.legend()
```

## Early Stopping 

```python
from sklearn.metrics import mean_squared_error
modelo= GradientBoostingRegressor(max_depth=2,n_estimators=200,random_state=42)

X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=49)

modelo.fit(X_train,y_train)

mse=[mean_squared_error(y_val,y_pred) for y_pred in modelo.staged_predict(X_val)]

best_n_estimadores=np.argmin(mse)

best_modelo= GradientBoostingRegressor(max_depth=2,n_estimators=best_n_estimadores,random_state=42)
```

```python
best_modelo.fit(X,y)
plt.figure(figsize=(12,4))

plt.subplot(121)
plt.plot(np.arange(200),mse,label='mse')
plt.scatter(best_n_estimadores,mse[best_n_estimadores],label='mínimo',color='red')
plt.xlabel('Número de estimadores')
plt.ylabel('MSE')
plt.legend()

plt.subplot(122)
plt.scatter(X,y)
plt.plot(x1,best_modelo.predict(x1),color='red',label='predicción')
plt.title(f'Estimadores: {best_n_estimadores}')
plt.legend()

```

### En la gráfica izquierda se observa el comportamiento del error del modelo con respecto al número de estimadores del modelo, a su vez se indica el número de estimadores para el cual se produce el error mínimo (punto rojo). En la figura derecha se observa el modelo entrenado con el parametro mencionado anteriormente.

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(mnist.data,mnist.target,test_size=10000)
X_train,X_val,y_train,y_val= train_test_split(X_train,y_train,test_size=10000)
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.neural_network import MLPClassifier

rf= RandomForestClassifier(n_jobs=-1,n_estimators=200,random_state=42)
etc=ExtraTreesClassifier(n_jobs=-1,n_estimators=200,random_state=42)
mlpc=MLPClassifier(random_state=42)

for modelo in [rf,etc,mlpc]:
    modelo.fit(X_train,y_train)
    print(modelo.__class__.__name__,": ",modelo.score(X_val,y_val))
```

```python
from sklearn.ensemble import VotingClassifier
lista=[('rf',rf),('etc',etc),('mlpc',mlpc)]
vt=VotingClassifier(estimators=lista)
vt.fit(X_train,y_train)
vt.score(X_val,y_val)
```

### Se observa que el rendiemiento de VotingClassifier es similar al obtenido por uno de sus estimadores el cual es ExtraTreeClassifier. Para este caso en especial ExtraTreeClassifier es mejor ya que tiene un rendimiento calcado y además es significativamente más rápido a la hora de entrenarlo en comparación al ensemble.

```python
etc.score(X_test,y_test)
```

```python
[modelo.score(X_test,y_test) for modelo in [rf,etc,mlpc,vt]]
```

### En general se observa muy buen rendimiento de cada uno de los estimadores. Ironicamente Random forests y ExtraTree tienen mejor predicción  que el ensemble estimador sin contar que son mucho más rápido en el proceso de fitting. Por lo tanto para este caso el modelo a escoger sería el ExtraTreeClassifier.
